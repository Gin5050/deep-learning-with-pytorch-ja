```python
import torch
input_g = torch.tensor([[[1, 2, 3, 4, 3], [4, 5, 7, 3, 1], [2, 1, 6, 0, 4], [2, 3, 4, 1, 1], [2, 3, 4, 1, 1]], [[1, 2, 3, 4, 3], [4, 5, 7, 3, 1], [2, 1, 6, 0, 4], [2, 3, 4, 1, 1], [2, 3, 4, 1, 1]]])
```

```python
input_g.shape
```

```python
input_g.unsqueeze_(0)
```

```python
input_g.to(torch.float32)
```

```python
transform_t = torch.eye(3)
```

```python
for i in range(2):
    transform_t[2, i] = 2
```

```python
transform_t
```

```python
transform_t = transform_t.expand(input_g.shape[0], -1, -1)
```

```python
transform_t.shape
```

```python
input_g.shape
```

```python
transform_t = transform_t.to(input_g.device, torch.float32)
```

```python
import torch.nn.functional as F
affine_t = F.affine_grid(transform_t[:, :2], input_g.size(), align_corners=False)
```

```python
augmented_input_g = F.grid_sample(input_g, affine_t, padding_mode="border", align_corners=False)
```
