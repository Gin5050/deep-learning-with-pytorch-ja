```python
def run(app, *argv):
    argv = list(argv)
    argv.insert(0, '--num-workers=4')  # <1>
    log.info("Running: {}({!r}).main()".format(app, argv))

    app_cls = importstr(*app.rsplit('.', 1))  # <2> 引数のモジュールをimport して返す?
    app_cls(argv).main() # パーサーで条件を付与して, main関数を実行する?

    log.info("Finished: {}.{!r}).main()".format(app, argv))
```
